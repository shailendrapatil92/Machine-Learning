# Import Statements

```python
import pandas as pd
import numpy as np
import scipy.io as sio
import os
```

```python
vocabDict = {}
X_train = None
Y_train = None
weights = []
```

# Import topics data

```python
data_topics_csv = pd.read_csv("reuters/all-topics-strings.lc.txt", header = None)
data_topics = np.matrix(data_topics_csv)

for ptr in range(0, data_topics.shape[0]):
    data_topics[ptr, 0] = data_topics[ptr, 0].strip()
```

# Import Vocab Consolidated

```python
vocab_output = open('vocab_main.csv', 'r')
vocabList = vocab_output.readline().replace('\r\n','')
vocabList = vocabList.split(',')

print len(vocabList)
```

# Create a Dictionary from the Consolidated Vocab

```python
for ptr in range(0, len(vocabList)):
    vocabDict.update({vocabList[ptr]: ptr})
    
print len(vocabDict)
```

# Parse all the documents and create Train data

```python
input_dir = os.listdir('files')
for f in input_dir:
    input_file = open('files/'+str(f),'r')
    print "Parsing File: " + str(f)
    
    input_file_data = input_file.readline().replace('\r\n','')
    input_file_data = input_file_data.split(',')
    #print input_file_text
    
    Y = input_file.readline().replace('\r\n','').replace('[','').replace('.]','').strip()
    Y = Y.split(',')
    
    Y = map(float, Y)
    Y = map(int, Y)
    Y = np.matrix(Y)
    
    for x in range(0, len(input_file_data)):
        if vocabDict.get(input_file_data[x]) != None:
            weights += [vocabDict.get(input_file_data[x])]
            
    X = np.zeros((len(vocabDict), 1))
    
    if len(weights) > 1:
        X[np.matrix(weights), 0] = 1
        
    if X_train == None:
        X_train = np.matrix(X.reshape((1, X.shape[0])))
    else:
        X_train = np.append(X_train, X.reshape((1,X.shape[0])), axis=0)
        
    if Y_train == None:
        Y_train = np.matrix(Y)
    else:
        Y_train = np.append(Y_train, Y, axis=0)
    
print "X_train shape: ", X_train.shape
print "Y_train shape: ", Y_train.shape
```

# Save train data to local

```python
sio.savemat('X_train.mat', {'X_train': X_train})
sio.savemat('Y_train.mat', {'Y_train': Y_train})
```
